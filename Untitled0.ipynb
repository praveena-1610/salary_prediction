```python
# we are  importing pandas to manuplate data
import pandas as pd
# we importing matplotlib for visual data
import matplotlib.pyplot as plt
# by sklearn.linear_model we are inporting linearRegression
from sklearn.linear_model import LinearRegression
```

```python
# here we are reading the data it will gave all data in file
pd.read_csv("/content/Salary_dataset.csv")
```

```python
# we are droping the column we don't need
pd.read_csv("/content/Salary_dataset.csv").drop("Unnamed: 0",axis=1)
```

```python
# we are giviing the variable data
data = pd.read_csv("/content/Salary_dataset.csv").drop("Unnamed: 0",axis=1)
```

```python
# we are just getting data
data
```

```python
#  want just one column that we are doing here
data["Salary"]
```

```python
# here we are assing x and y values
x=data["YearsExperience"]
y=data["Salary"]
```

```python
x
```

```python
# we are pioting the graph
plt.scatter(x,y)
```

instalissing

```python
# we are instalissing means we are creating a object for our class
model = LinearRegression()
```

```python
# we are calling the linear regression
model.fit(pd.DataFrame(x),y)
```

```python
#we are predicting the data
model.predict([[6]])
```

```python
# we are pioting the linear line in the graph
plt.scatter(x,y)
plt.plot(x,model.predict(pd.DataFrame(x)))
```

```python
# getting y-intercept
c=model.intercept_
```

```python
# getting slope
m=model.coef_
```

```python
# caluclating y value
y = m * 3 + c
y
```
